```python
import numpy as np #Para crear vectores y arreglos
import random #Para obtener números aleatorios
from sklearn import linear_model as lm #Para utilizar el modelo de regresión lineal
from sklearn.metrics import mean_squared_error, r2_score #Para calcular métricas y predicciones
import matplotlib.pyplot as plt
 
# Generador de distribución de datos para regresión lineal simple
def generador_datos_simple(beta, muestras, desviacion):
    x = np.random.random_sample(muestras) * 100 # Generación de n (muestras) valores de x aleatorios entre 0 y 100)
    e = np.random.randn(muestras) * desviacion # Generación un error aleatorio gaussiano con desviación típica (desviacion)
    y = x * beta + e # Obtener el y real como x*beta + error
    return x.reshape((muestras,1)), y.reshape((muestras,1))
 
# Parámetros de la distribución (dados por nosotros)
desviacion = 200 #dispersión de los datos
beta = 10 #beta sub 1
n = 50 #número de muestras aleatorias
x, y = generador_datos_simple(beta, n, desviacion)
 
# Los datos generados graficamente
plt.scatter(x, y) #Creación de un diagrama de dispersión
plt.show() #Mostrar el diagrama de dispersión creado
```

Teniendo los datos creados

```python
modelo = lm.LinearRegression() # Creación de un modelo de regresión lineal
 
modelo.fit(x,y) # Ajustar el modelo con los datos (X,Y)
print ('Coeficiente beta1: ', modelo.coef_[0]) # Coeficiente B0
 
y_pred = modelo.predict(x) # Podemos predecir usando el modelo
 
print ('Error cuadrático medio: %.2f' % mean_squared_error(y, y_pred)) # Por último, calculamos el error cuadrático medio
print ('Estadístico R_2: %.2f' % r2_score(y, y_pred))  # y calculamos el estadístico R^2
```

Creando líneas de tendencia

```python
plt.scatter(x, y) #Creación de un diagrama de dispersión
plt.plot(x, y_pred, color='red') #Creación de una línea de tendencia de predicción en color rojo
x_real = np.array([0, 100]) #Creación de un arreglo
y_real = x_real*beta #Obtención de coordenadas Y del arreglo
plt.plot(x_real, y_real, color='green') #Creación de una linea de tendencia real en color verde
plt.show() #Mostrar gráfico creado
```
